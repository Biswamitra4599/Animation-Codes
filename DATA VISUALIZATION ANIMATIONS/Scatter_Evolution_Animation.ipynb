## SCATTER EVOLUTION — FULL WORKING CODE

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter

# -----------------------------------------
# Output folder
# -----------------------------------------
os.makedirs("gifs", exist_ok=True)

# -----------------------------------------
# Generate evolving data
# -----------------------------------------
np.random.seed(42)

n_points = 80
n_steps = 60

# Initial positions
X = np.random.randn(n_points, 2)

# Velocities (motion direction)
velocity = np.random.randn(n_points, 2) * 0.05

# Store trajectory
history = []

for t in range(n_steps):
    X = X + velocity
    history.append(X.copy())

# -----------------------------------------
# Plot setup
# -----------------------------------------
fig, ax = plt.subplots(figsize=(6, 6))

scatter = ax.scatter([], [], s=40, c="dodgerblue", alpha=0.7)

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_title("Scatter Evolution Animation")

# -----------------------------------------
# Animation
# -----------------------------------------
writer = PillowWriter(fps=6)

with writer.saving(fig, "gifs/scatter_evolution.gif", dpi=120):

    for i in range(n_steps):
        scatter.set_offsets(history[i])
        ax.set_title(f"Scatter Evolution — Step {i+1}")

        writer.grab_frame()

plt.show()
```
