```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Generate sample data
np.random.seed(0)
X = np.linspace(0, 10, 50)
y = 2.5 * X + 5 + np.random.randn(50) * 4

# Gradient descent
m, b = 0, 0
lr = 0.001
iterations = 200

m_history, b_history = [], []

for _ in range(iterations):
    y_pred = m * X + b
    error_m = (-2/len(X)) * np.sum((y - y_pred) * X)
    error_b = (-2/len(X)) * np.sum(y - y_pred)

    m -= lr * error_m
    b -= lr * error_b

    m_history.append(m)
    b_history.append(b)

# Animation
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(X, y, color="blue", label="Data Points")
line, = ax.plot([], [], color="red", linewidth=2)

ax.set_xlim(0, 10)
ax.set_ylim(min(y)-5, max(y)+5)
ax.set_title("Linear Regression - Gradient Descent Learning")
ax.legend()

def update(i):
    y_pred = m_history[i] * X + b_history[i]
    line.set_data(X, y_pred)
    ax.set_title(f"Linear Regression Learning - Step {i+1}")
    return line,

ani = FuncAnimation(fig, update, frames=iterations, interval=40)

ani.save("gifs/linear_regression.gif", writer="pillow")
```
